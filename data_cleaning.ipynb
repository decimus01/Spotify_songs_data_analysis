## Selection by performers
### Where there is more than one performer, they will be separated so that there is only one performer in each row.

### Temporary dataframe

```python
tmp = pd.DataFrame()
```

### Separate artists and store them in a list in the dataframe

```python
tmp = songlist
tmp['artists'] = tmp['artists'].str.split(',').apply(lambda x: [artists.strip() for artists in x])
```

### One performer in each row

```python
df_songlist = tmp.explode('artists')
df_songlist.head()
```

## Cleaning

### Replace NaN with 'empty'

### Replace 'empty' with 'Global Top 50'

```python
#"The ISO code of the country of the Top 50 Playlist. If Null, then the playlist if 'Global Top 50'." from Kaggle

songdailyrank.loc[songdailyrank_index, 'country'] = 'Global Top 50'
songdailyrank[songdailyrank.country == 'Global Top 50']
```

## Drop unknown songs from table
